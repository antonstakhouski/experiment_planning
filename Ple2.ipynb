```python
# The usual preamble
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import normalize
import scipy.stats

# Make the graphs a bit prettier, and bigger
plt.show(block=True)
plt.style.use('bmh')

plt.rcParams['figure.figsize'] = (15, 5)
plt.rcParams['font.family'] = 'sans-serif'
```

Загружаем датасет.

```python
dataset = pd.read_csv("athlete_events.csv")[:20000]
dataset[:10]
```

```python
df = pd.DataFrame({'Age': dataset['Age'], 'Height': dataset['Height']})
table = df.dropna()
table[:5]
```

```python
len(table)
```

```python
table = table.sort_values(by=['Age'])
table
df = pd.DataFrame({'Age rank' :table['Age'].rank()}, index=table.index)
table = table.add(df, fill_value=0)
df = pd.DataFrame({'Height rank' :table['Height'].rank()}, index=table.index)
table = table.add(df, fill_value=0)
```

```python
table[:10]
```

```python
table = table.reset_index(drop=True)
table[:10]
```

```python
arr = [np.where(table['Height rank'][i:] 
                                     > val)[0].size for i, val in 
enumerate(table['Height rank'])]
df = pd.DataFrame({'P' :arr}, index=table.index)
table = table.add(df, fill_value=0)

arr = [np.where(table['Height rank'][i:]
                                     < val)[0].size for i, val in 
enumerate(table['Height rank'])]
df = pd.DataFrame({'Q' :arr}, index=table.index)
table = table.add(df, fill_value=0)
table[:10]
```

```python
p = sum(table['P'])
p
```

```python
q = sum(table['Q'])
q
```

```python
n = len(table)
n
```

```python
t = (p - q) / (1/2 * n * (n - 1))
t
```

Коэффициент Кедналла: корреляционная связь несущественна

```python
arr = [(table['Age rank'][i] - table['Height rank'][i]) ** 2
 for i in table.index.values]
df = pd.DataFrame({'d ^ 2' :arr}, index=table.index)
table = table.add(df, fill_value=0)
table[:10]  
```

```python
p = 1 - 6 * sum(table['d ^ 2']) / (n ** 3 - n)
p
```
